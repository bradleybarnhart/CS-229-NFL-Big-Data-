```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

scores = pd.read_csv('nfl-big-data-bowl-2020-publicleaderboard.csv')
scores = scores.sort_values(by = 'Score', ascending = False).reset_index()
scores = scores[40:]

sns.set(font_scale=2)
plt.plot(scores['Score'], '-o') 
plt.plot(scores[scores['TeamName']=='229']['Score'], 'o', markersize = 18)
plt.title('Team Performance after Week 1')
plt.xlabel('TeamID')
plt.ylabel('Score (cost)')

print('performance comparison (disregarding first 40 teams)')
```
